#2. Data Structure: The List Container
Lists are collections of items that are ordered and mutable (meaning you can change them later).

```python

# --- 2. Data Structures: Lists (Ordered Container) ---

# Define a list of items
fruits = ["apple", "banana", "cherry"] [2]

# Accessing an item by position (Lists start counting from zero) [2]
# Index 1 refers to the second item: "banana"
print(fruits[2])
# Expected Output: banana
```

#3. Control Flow: The for Loop
The for loop is used when you want to perform an action for every item in a collection.

```python

# --- 3. Control Flow: The for Loop (Iteration) ---

# List of numbers to iterate over [3]
numbers = [1, 2, 3, 4, 5]
total = 0 # Initialize a running total at zero [3]

# Loop runs for every number in the list
# It grabs the 1, adds it, grabs the 2, adds it, and so on [3]
for num in numbers:
    total = total + num # Add the current number to the running total

# Prints the final result (1 + 2 + 3 + 4 + 5 = 15) [3]
print(total)
# Expected Output: 15
```

#4. Control Flow: The while Loop
The while loop keeps running only as long as a specified condition remains True.

```python


# --- 4. Control Flow: The while Loop (Conditional Repetition) ---

count = 3 # Start with count set to three [4]

# The loop rule: Keep running while count is greater than zero [4]
while count > 0:
    print(count) # Print the current count

    # Crucial step: We must subtract one from count, otherwise the loop runs forever [4]
    count = count - 1

# This line runs once the condition (count > 0) is finally broken (when count becomes 0) [4]
print("Blastoff!")
# Expected Output: 3, 2, 1, Blastoff!
```

#5. Functions and The Final Challenge
Functions help organize code into reusable blocks, each intended to have one and only one job. This final example combines all core concepts (data, lists, loops, and conditional logic) into a function

```python
# Define a function that takes a list of numbers as input [5]
def sum_even_numbers(number_list):

    # Create a variable for our sum, starting at zero [5]
    even_sum = 0

    # Loop through every number provided in the list [6]
    for number in number_list:

        # Check if the number is even using the modulo operator (%) [6]
        # Modulo gives the remainder of a division. If remainder when divided by 2 is 0, it's even.
        if number % 2 == 0:
            even_sum += number # If it's even, add it to our running sum [6]

    # Return the final sum when the loop has finished checking every number [6]
    return even_sum
```

```python



# Test Data: The function should find 2, 4, and 6.
data = [1, 2, 3, 4, 5, 6,10]

# Call the function and store the result
result = sum_even_numbers(data)

# Display the final sum (2 + 4 + 6 = 12) [6]
print(result)
# Expected Output: 12
```
